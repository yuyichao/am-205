```julia
include("common.jl")
using PSet1Common
```

```julia
xs = [0, 1, 2, 3];
ys = [0, 0, 1, 2];
```

```julia
monomial_res = solve_interpolation(MonomialBase(), xs, ys)
```

```julia
lagrange_res = solve_interpolation(LagrangeBase(xs), xs, ys)
```

```julia
xs = linspace(0, 3, 1000)
mono_ys = monomial_res(xs) - 0.01
lagr_ys = lagrange_res(xs) + 0.01

plot(xs, mono_ys, label="Monomial")
plot(xs, lagr_ys, label="Lagrange")
legend()
grid()
```
