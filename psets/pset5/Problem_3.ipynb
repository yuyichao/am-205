```julia
include("eig.jl")
```

```julia
immutable V1
end
call(::V1, x) = abs(x)
immutable V2
end
call(::V2, x) = @evalpoly(x, 13 / 10, 1 / 8, -1 / 18, 0, 12 / 10^4)
immutable V3
end
call(::V3, x) = 8 * abs(abs(abs(x) - 1) - 1)
```

```julia
xs = linspace(-12, 12, 1921)
```

```julia
Ds1, Vecs1 = eigs_ode(V1(), xs)
plot_eigs(V1(), xs, Ds1, Vecs1)
```

```julia
Ds2, Vecs2 = eigs_ode(V2(), xs)
plot_eigs(V2(), xs, Ds2, Vecs2)
```

```julia
Ds3, Vecs3 = eigs_ode(V3(), xs)
plot_eigs(V3(), xs, Ds3, Vecs3)
```
